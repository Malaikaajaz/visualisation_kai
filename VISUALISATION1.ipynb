```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
Path = '/content/drive/MyDrive/KollectAi/Datasets/FinalDataset.csv'
df= pd.read_csv(Path)

```

```python
# Filter the DataFrame for total arrears greater than 0.0
filtered_df = df[df['total_arrears'] > 0.0]

# Calculate the total arrears percentage for each gender
gender_total_arrears = filtered_df.groupby('gender_description')['total_arrears'].mean()
total_arrears_percentage = (gender_total_arrears / gender_total_arrears.sum()) * 100

# Create a DataFrame for plotting
data = pd.DataFrame({'Gender': total_arrears_percentage.index, 'Total Arrears Percentage': total_arrears_percentage.values})

# Sort the data by 'Total Arrears Percentage' in ascending order
data = data.sort_values('Total Arrears Percentage')

# Plot the bar chart with hover effect and pastel colors
fig = px.bar(data, x='Gender', y='Total Arrears Percentage', color='Gender', hover_data=['Total Arrears Percentage'],
             labels={'Total Arrears Percentage': 'Total Arrears Percentage (%)'},
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_layout(title='Total Arrears Percentage by Gender', xaxis_title='Gender', yaxis_title='Total Arrears Percentage (%)')

fig.show()
```

```python
# Convert the 'disbursement_date' column to a datetime data type if needed
df['disbursement_date'] = pd.to_datetime(df['disbursement_date'])

# Create a new column 'disbursement_month_year' as string
df['disbursement_month_year'] = df['disbursement_date'].dt.to_period('M').astype(str)

# Group the data by 'disbursement_month_year' and calculate the default rate for each group
default_rates = df.groupby('disbursement_month_year')['customer_status_description'].apply(lambda x: (x == 'Default').mean() * 100)

# Plot the default rates over time
fig = px.line(default_rates, x=default_rates.index, y=default_rates.values)
fig.update_layout(title='Default Rate by Disbursement Month/Year', xaxis_title='Disbursement Month/Year', yaxis_title='Default Rate (%)')
fig.show()
```

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap using Plotly Express
fig = px.imshow(correlation_matrix)
fig.update_layout(title='Correlation Matrix')
fig.show()
```

```python
# Filter the DataFrame for relevant columns: 'gender_description' and 'total_arrears'
gender_arrears = df[df['total_arrears'] > 0.0][['gender_description', 'total_arrears']]

# Calculate the percentage of total arrears for each gender
gender_arrears['arrears_percentage'] = gender_arrears.groupby('gender_description')['total_arrears'].transform(lambda x: x / x.sum() * 100)

# Create a scatter plot
fig = px.scatter(gender_arrears, x='gender_description', y='arrears_percentage', color='gender_description',
                 hover_data=['arrears_percentage'], labels={'gender_description': 'Gender', 'arrears_percentage': 'Total Arrears (%)'})

# Set plot aesthetics
fig.update_layout(title='Gender vs. Total Arrears Percentage',
                  xaxis_title='Gender', yaxis_title='Total Arrears (%)',
                  plot_bgcolor='rgba(240, 240, 240, 0.5)', paper_bgcolor='rgba(240, 240, 240, 0.5)',
                  font=dict(family='Arial', size=12, color='black'))

# Set pastel color palette
pastel_palette = px.colors.qualitative.Pastel

# Update marker colors
fig.update_traces(marker=dict(color=pastel_palette))

# Show the plot
fig.show()
```
